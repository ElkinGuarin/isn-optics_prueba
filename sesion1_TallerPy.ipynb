```python
nm=1e-9
um=1e-6
mm=1e-3
import matplotlib.pyplot as plt

#longitud de onda 650 nm

wave_lenght=650*nm
k=2*np.pi/wave_lenght

wave_range=np.arange(0,5*um,10*nm)
onda1D=np.sin(k*wave_range)

plt.plot(onda1D)
plt.show()
```

```python


FFT_onda1D=np.fft.fft(onda1D)

#indexacion, posiciones entre 450 y 550
plt.plot(
    (np.abs(
        np.fft.fftshift(FFT_onda1D)
    )**2)[450:550],'+'
    )
```

```python
#Creemos una definicion

def Plane_wave1D(w_lenght,phase,Range_1D):
  k=2*np.pi/w_lenght
  return np.exp(1j*(k*Range_1D)+phase)
  
```

```python
Complex_PW1D=Plane_wave1D(wave_lenght,0,wave_range)
```

```python
plt.plot(np.real(Complex_PW1D))
plt.plot(np.imag(Complex_PW1D))
plt.plot(np.abs(Complex_PW1D))
plt.plot(np.angle(Complex_PW1D))
plt.show()
```
